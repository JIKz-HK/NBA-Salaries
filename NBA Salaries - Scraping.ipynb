```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
#loop for all pages and convert table to dataframe
df = pd.DataFrame()
url_loop = 'http://www.espn.com/nba/salaries/_/page/'
for page in range(1,13):
    webpage = requests.get(url_loop + str(page) + '/')
    soup = BeautifulSoup(webpage.text,'html.parser')
    table = soup.find_all('table')
    temp_df = pd.read_html(str(table))
    df = pd.concat([df, temp_df[0]])
```

```python
df
```

```python
#data cleaning
df = df.dropna()
df = df.drop_duplicates()
df.columns = ['Rank', 'Name', 'Team', 'Salary']
df = df.drop(df.loc[df['Rank'].str.contains('RK')].index)
```

```python
#dataframe sort
df['Salary'] = df['Salary'].str.replace(',', '')
df['Salary'] = df['Salary'].str.replace('$', '')
df['Salary'] = df['Salary'].astype(int)
df['Rank'] = df['Rank'].astype(int)
df = df.sort_values(by=['Rank'])
df[['Name','Position']] = df['Name'].str.split(",",expand=True)
df['Position'] = df['Position'].str.strip()
```

```python
#index reset
df = df.reset_index().drop(columns = 'index')
```
